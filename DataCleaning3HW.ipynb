### Q2 ###

Notice that the data has a time dimension (year). For this assignment, forget about time and treat all the observations as if they're from the same year. Basically all of these would make sense to fill in with a measure of central tendency, but the last four (math and reading scores) are probably missing too much data to be meaningful. Thsese should probably just be tossed. 

```python
df2 = df.copy()
fill_cols = [
    "ENROLL",
    "TOTAL_REVENUE",
    "FEDERAL_REVENUE",
    "STATE_REVENUE",
    "LOCAL_REVENUE",
    "TOTAL_EXPENDITURE",
    "INSTRUCTION_EXPENDITURE",
    "SUPPORT_SERVICES_EXPENDITURE",
    "OTHER_EXPENDITURE",
    "CAPITAL_OUTLAY_EXPENDITURE",
    "GRADES_PK_G",
    "GRADES_KG_G",
    "GRADES_4_G",
    "GRADES_8_G",
    "GRADES_12_G",
    "GRADES_1_8_G",
    "GRADES_9_12_G",
    "GRADES_ALL_G",
]
for col in fill_cols:
    df2.loc[:, col] = df2.loc[:, col].fillna(df2.loc[:, col].mean())
df2
```

### Q3 ###

Now, take into account the time factor. Replicate your second answer but this time fill in the missing values by using a statistic that is calculated within the year of the observation. For example, if you want to fill a missing value for a variable with the mean of that variable, calculate the mean by using only the observations for that specific year.


```python
df3 = df.copy()

# for col in fill_cols:
#     df2.loc[:, col] = df2.loc[:, col].fillna(
#         df2.loc[:, [col, "YEAR"]].groupby(by=["YEAR"]).mean()
#     )

# for col in fill_cols:
#     df3[col] = df3[col].fillna(df3.groupby(by="YEAR").mean())

for col in fill_cols:
    df3[col] = df3.groupby(["YEAR"], sort=False)[col].apply(
        lambda x: x.fillna(x.mean())
    )
```

### Q4 ###

This time, fill in the missing values using interpolation (extrapolation).


```python
df4 = df.copy()

for col in fill_cols:
    df4[col] = df4[col].interpolate()
```

```python
for col in fill_cols:
    print("Statistics for columns: {}".format(col))
    print(pd.concat([df[col], df2[col], df3[col], df4[col],], axis=1,).describe())
```

Seems like, for the grades,  interpolation generally reduces the median, while it increases it for measures of money. This probably has more to do with the ordering of the data than anything.

Importantly, the overall median is almost always higher than the year by year one. I suspect that this is because you have more data today, when both spending and grades are higher. So you're filling in more (lower) values for earlier dates.
